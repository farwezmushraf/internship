```python
ar1 = np.array([1,2,3,4,5])
ar1
```

```python
np.mean(ar1)
```

```python
#mean function
def mean(array):
    n=len(array)
    return np.sum(array)/n
mean(ar1)
```

```python
#variance function:
def variance(array):
    n = len(array)
    m = mean(array)
    return np.sum((array -m )**2)/n
variance(ar1)
```

```python
#min-max scalar
def minmax(array):
    min = np.min(array)
    max = np.max(array)
    return ((array) - min) / (max - min)
minmax(ar1)
```

```python
#standard scalar:
def std_sc(array):
    m = mean(array)
    s = variance(array)
    return (array - m) / s
std_sc(ar1)
```

```python
#sigmoid:
def sigmoid(array):
    exp = np.exp(-array)
    return 1 / 1 + exp
sigmoid(ar1)
```

```python
#relu:
def relu(array):
    return np.maximum(0,array)  
relu(ar1)
```
