```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier , RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier 
from sklearn.preprocessing import StandardScaler ,LabelEncoder
```

```python
df = pd.read_csv(r"C:\Users\farwee\OneDrive\Desktop\Titanic-Dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.columns
```

```python
df.drop(columns=['PassengerId','Name','Ticket'],inplace = True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.Age.fillna(df.Age.mean(),inplace=True) 
```

```python
df.isnull().sum()
```

```python
df.Cabin.fillna(value = df.Cabin.mode()[0].split()[1], inplace = True)
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
encoder = LabelEncoder()
df.Sex = encoder.fit_transform(df.Sex)
df.Cabin = encoder.fit_transform(df.Cabin)
df.Embarked = encoder.fit_transform(df.Embarked)
```

```python
df.head()
```

```python
scaler = StandardScaler()
df.Age = scaler.fit_transform(df[['Age']])
```

```python
df.head()
```

```python
x = df.drop(columns='Survived')
y= df.Survived
```

```python
x.head()
```

```python
x_train, x_test,y_train, y_test = train_test_split(x,y, random_state = 42,test_size=0.2)
```

```python
base_model = DecisionTreeClassifier(max_depth = 2, random_state = 42)
```

```python
base_model
```

```python
classifier = AdaBoostClassifier(random_state = 42,estimator = base_model, n_estimators = 100)
```

```python
classifier
```

```python
classifier.fit(x_train, y_train)
```

```python
pred = classifier.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix, precision_score, recall_score
```

```python
confusion_matrix(pred, y_test)
```

```python
accuracy_score(pred, y_test)
```

```python
precision_score(pred, y_test)
```

```python
recall_score(pred,y_test)
```
